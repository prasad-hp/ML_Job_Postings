```python
df['job_title'].value_counts().head(10).plot(kind='bar')
```

```python
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])

df['job_posted_date'].value_counts().sort_index().plot()
plt.title("Job Postings Over Time")
plt.xlabel("Date")
plt.ylabel("Number of Postings")
plt.grid(True)
```

```python
plt.figure(figsize=(10, 5))
df['company_address_region'].value_counts().head(15).plot(kind='bar')
plt.title("Jobs by Region")
plt.ylabel("Number of Jobs")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
df['company_name'].value_counts().head(10).plot(kind='bar')
plt.title("Top 10 Companies by Job Posts")
plt.xticks(rotation=45)
plt.show()
```
